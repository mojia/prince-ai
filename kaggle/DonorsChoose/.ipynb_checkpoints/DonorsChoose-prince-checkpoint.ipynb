```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.model_selection import cross_validate
from sklearn.model_selection import ShuffleSplit
from sklearn.svm import SVR
from sklearn.svm import SVC



dir='/Users/xinwang/ai/dataset/kaggle/DonorsChoose/'
train_file='train.csv'
test_file='test.csv'
resource_file='resources.csv'
SEED=1000


train_df = pd.read_csv(dir + train_file)
test_df = pd.read_csv(dir + test_file)
resource_df = pd.read_csv(dir + resource_file)
label = LabelEncoder()

def sampleData():
    train_label_1_df = train_df[train_df['project_is_approved']==1].sample(n=30000,random_state=SEED)
    train_label_0_df = train_df[train_df['project_is_approved']==0]

    train_data = pd.concat([train_label_1_df,train_label_0_df])
    train_data = shuffle(train_data)
    
    return train_data

def process_feature_teacher_prefix():
    teacher_prefix_mapping = {'Mrs.':1, 'Ms.':2, 'Mr.':3,'Teacher':4, 'Dr.':5}
    train_data['teacher_prefix'] = train_data['teacher_prefix'].map(teacher_prefix_mapping)

    train_data['teacher_prefix'] = label.fit_transform(train_data['teacher_prefix'])


def process_feature_school_state():
    train_data['school_state'] = label.fit_transform(train_data['school_state'])


def process_feature_project_grade_category():
    train_data['project_grade_category'] = label.fit_transform(train_data['project_grade_category'])
    
    
def process_feature_project_subject_categories():
    train_data['project_subject_categories'] = label.fit_transform(train_data['project_subject_categories'])
    
def process_feature_project_subject_subcategories():
    train_data['project_subject_subcategories'] = label.fit_transform(train_data['project_subject_subcategories'])
    
def process_feature_teacher_number_of_previously_posted_projects():
    train_data['teacher_number_of_previously_posted_projects_code'] = pd.cut(
        train_data['teacher_number_of_previously_posted_projects'], bins=4)

    train_data['teacher_number_of_previously_posted_projects'] = label.fit_transform(
        train_data['teacher_number_of_previously_posted_projects_code'])
    
    train_data.drop(['teacher_number_of_previously_posted_projects_code'], axis=1, inplace=True)
    
def process_feature_teacher_id():
    train_data['teacher_id'] = label.fit_transform(
        train_data['teacher_id'])
    
def process_feature_project_submitted_datetime():
    train_data['project_submitted_year']  = train_data['project_submitted_datetime'].str.split(
        ' ', expand=True)[0].str.split('-', expand=True)[0]
    train_data['project_submitted_year'] = label.fit_transform(train_data['project_submitted_year'])
    
    
    train_data['project_submitted_month'] = train_data['project_submitted_datetime'].str.split(
        ' ', expand=True)[0].str.split('-', expand=True)[1]
    
    train_data['project_submitted_day'] = train_data['project_submitted_datetime'].str.split(
        ' ', expand=True)[0].str.split('-', expand=True)[2]
    train_data.drop(['project_submitted_datetime'], axis=1, inplace=True)
    

def drop_columns():
    drop_columns = ['id','teacher_id','project_submitted_datetime','project_title',
                    'project_essay_1','project_essay_2','project_essay_3','project_essay_4',
                   'project_resource_summary']
    train_data.drop(drop_columns, axis=1, inplace=True)
    
    
train_data = sampleData()

process_feature_teacher_prefix()
process_feature_school_state()
process_feature_project_grade_category()
process_feature_project_subject_categories()
process_feature_project_subject_subcategories()
process_feature_teacher_number_of_previously_posted_projects()
process_feature_teacher_id()
process_feature_project_submitted_datetime()


cv_split = ShuffleSplit(n_splits=5, test_size=0.3, train_size=0.6, random_state=SEED)


features= ['teacher_prefix','school_state','project_grade_category',
           'project_subject_categories','project_subject_subcategories','teacher_number_of_previously_posted_projects',
          'teacher_id','project_submitted_year','project_submitted_month','project_submitted_day']
target = 'project_is_approved'


train_data_x= train_data[features]
train_data_y = train_data[target].ravel()
```

```python
def correlation_heatmap(df):
    _,ax = plt.subplots(figsize=(14,12))
    
    colormap = sns.diverging_palette(220, 10, as_cmap=True)
    _ = sns.heatmap(df.corr(),
                   cmap = colormap,
                   square=True,
                   cbar_kws={'shrink':.9},
                   ax=ax,
                   annot=True,
                   linewidths=0.1,
                    vmax=1.,
                    linecolor='white')
    plt.title('Pearson correlation of features', y=1.05, size=15)

    
correlation_heatmap(train_data)  
```

```python
print('SVR start ...')
model = SVC(probability=True)
cv_results = cross_validate(model, train_data_x, train_data_y, cv=cv_split, scoring='roc_auc', verbose=2)
print('train accuracy mean:', cv_results['train_score'].mean()*100.)
print('test accuracy mean:', cv_results['test_score'].mean()*100.)
print('test accuracy 3*std', cv_results['test_score'].std()*3*100.)
```
